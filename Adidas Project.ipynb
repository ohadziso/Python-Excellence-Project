## **Minimum Price per Unit**

```python
#Assigning the desired results to a new variable
minimum_price_per_unit = adidas['Price per Unit'].min()

#Presenting the data in a more readable and clear format
int(minimum_price_per_unit)
```

```python
#Assigning the desired results to a new variable
unique_retailers = adidas['Retailer'].nunique()

#Displaying the received data
unique_retailers
```

## **Total Units Sold for each Product Gender**

```python
#Assigning the desired results to a new variable
total_units_sold_by_gender = adidas.groupby('Product Gender')['Units Sold'].sum().to_frame()

#Presenting the data in a more readable and clear format
total_units_sold_by_gender['Units Sold'].apply('{:,}'.format).sort_values(ascending=False).to_frame() 
```

## **Total Sales for each Product Gender**

```python
#Assigning the desired results to a new variable
total_sales_by_gender = adidas.groupby('Product Gender')['Total Sales'].sum().to_frame()

#Presenting the data in a more readable and clear format
total_sales_by_gender['Total Sales'].astype(int).apply('{:,}'.format).sort_values(ascending=False).to_frame()
```

## **Mean Operating Profit for each Region**

```python
#Assigning the desired results to a new variable
mean_operating_profit_by_region = adidas.groupby('Region')['Operating Profit'].mean().to_frame()

#Presenting the data in a more readable and clear format
mean_operating_profit_by_region['Operating Profit'].round(2).apply('{:,}'.format).sort_values(ascending=False).to_frame()
```

## **Maximum Price per Unit in each Sales Method category**

```python
#Assigning the desired results to a new variable
max_price_per_unit_by_sales_method = adidas.groupby('Sales Method')['Price per Unit'].max().to_frame()

#Presenting the data in a more readable and clear format
max_price_per_unit_by_sales_method.astype(int)
```

## **Total Total Sales for each combination of Product Gender and Region**

```python
#Assigning the desired results to a new variable
total_sales_by_gender_and_region = adidas.groupby(['Product Gender', 'Region'])['Total Sales'].sum().to_frame()

#Presenting the data in a more readable and clear format
total_sales_by_gender_and_region['Total Sales'].astype(int).apply('{:,}'.format).to_frame()
```

## **Mean Operating Margin for each Product**

```python
#Assigning the desired results to a new variable
mean_operating_margin_by_product = adidas.groupby('Product')['Operating Margin'].mean().to_frame()

#Presenting the data in a more readable and clear format
mean_operating_margin_by_product['Operating Margin'].sort_values(ascending=False).to_frame()
```

## **Maximum Total Sales in each State for Product Gender = Men**

```python
#Assigning the desired results to a new variable
max_total_sales_by_state_men = adidas[adidas['Product Gender'] == 'Men'].groupby('State')['Total Sales'].max().to_frame()

#Presenting the data in a more readable and clear format
max_total_sales_by_state_men['Total Sales'].astype(int).apply('{:,}'.format).sort_values(ascending=False).to_frame()
```

## **Maximum Total Sales in each State for Product Gender = Women**

```python
#Assigning the desired results to a new variable
max_total_sales_by_state_women = adidas[adidas['Product Gender'] == 'Women'].groupby('State')['Total Sales'].max().to_frame()

#Presenting the data in a more readable and clear format
max_total_sales_by_state_women['Total Sales'].astype(int).apply('{:,}'.format).sort_values(ascending=False).to_frame()
```

## **Mean Units Sold for each Region and Sales Method**

```python
#Assigning the desired results to a new variable
mean_units_sold_by_region_and_sales_method = adidas.groupby(['Region', 'Sales Method'])['Units Sold'].mean().to_frame()

#Presenting the data in a more readable and clear format
mean_units_sold_by_region_and_sales_method.round(3)
```

## **Total Operating Profit for each Product Gender, Region, and Sales Method**

```python
#Assigning the desired results to a new variable
total_operating_profit_by_gender_region_sales_method = adidas.groupby(['Product Gender', 'Region', 'Sales Method'])['Operating Profit'].sum().to_frame()

#Presenting the data in a more readable and clear format
total_operating_profit_by_gender_region_sales_method['Operating Profit'].astype(int).apply('{:,}'.format).sort_values(ascending=False).to_frame()
```

## **Median Price per Unit for each Product**

```python
#Assigning the desired results to a new variable
median_price_per_unit_by_product = adidas.groupby('Product')['Price per Unit'].median().to_frame()

#Presenting the data in a more readable and clear format
median_price_per_unit_by_product['Price per Unit'].astype(int).sort_values(ascending=False).to_frame()
```

## **Standard deviation of Total Sales for each Sales Method**

```python
#Assigning the desired results to a new variable
sales_method_total_sales_std = adidas.groupby('Sales Method')['Total Sales'].std().to_frame()

#Presenting the data in a more readable and clear format
sales_method_total_sales_std['Total Sales'].round(2).apply('{:,}'.format).sort_values(ascending=False).to_frame()
```

```python
#Comparing units sold to total sales for insights into product performance and revenue generation.

sns.scatterplot(x='Units Sold', y='Total Sales', data=adidas)
plt.title('Total Sales VS Units Sold')
plt.xlabel('Units Sold')
plt.ylabel('Total Sales')
ax=plt.gca()
ax.yaxis.set_major_formatter('{:,.0f}'.format)
plt.show()
```

```python
#Analyzing the relationship between unit price and operating profit for strategic insights and business optimization.

sns.scatterplot(x='Price per Unit', y='Operating Profit', data=adidas)
plt.title('Operating Profit VS Price per Unit')
plt.xlabel('Price per Unit')
plt.ylabel('Operating Profit')
ax=plt.gca()
ax.yaxis.set_major_formatter('{:,.0f}'.format)
plt.show()
```

```python
#Comparing invoice date with total sales for financial analysis and reconciliation purposes.

sns.lineplot(x='Invoice Date', y='Total Sales', data=adidas)
plt.title('Total Sales Trend Over Time')
plt.xlabel('Invoice Date')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
ax=plt.gca()
ax.yaxis.set_major_formatter('{:,.0f}'.format)
plt.show()
```

```python
#Summary of total sales across different regions,
#providing valuable insights for strategic decision-making and resource allocation.

sns.barplot(x='Region', y='Total Sales', data=adidas,errorbar=None, estimator=sum)
plt.title('Total Sales by Region')
plt.xlabel('Region')
plt.ylabel('Total Sales')
ax=plt.gca()
ax.yaxis.set_major_formatter('{:,.0f}'.format)
plt.show()
```

```python
#Accepting only numeric columns for analysis.

numerical_columns = ['Total Sales', 'Units Sold', 'Operating Profit', 'Price per Unit']

#Analyzing pairwise relationships: Total Sales, Units Sold, Operating Profit,
#and Price per Unit to draw valuable insights for data-driven business strategies.

sns.pairplot(adidas[numerical_columns])
plt.suptitle('Pairwise Relationships', y=1.02)
plt.gcf().set_size_inches(15, 13)
for ax in plt.gcf().axes:
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.show()
```

```python
#Exploring pairwise relationships and comparing them by product gender.

sns.pairplot(adidas, hue='Product Gender', vars=numerical_columns)
plt.suptitle('Pairwise Relationships with Product Gender Comparison', y=1.02)
for ax in plt.gcf().axes:
    ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
    ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
plt.show()
```

After analyzing the Adidas sales dataset, we have derived several key insights that can inform strategic decision-making and drive business success:

1. **Product Gender and Sales**: The highest sales amount is attributed to Men's street footwear. This indicates a strong market demand for this product category, presenting an opportunity for Adidas to focus on targeted marketing and product enhancements.

2. **Pricing Strategy**: The majority of products fall within the price range of `$40 to $60`, suggesting that this price bracket resonates well with consumers. A well-aligned pricing strategy can further capitalize on this sweet spot, maximizing revenue potential.

3. **Average Product Price**: The average price of products is `$45.22`. Understanding this average can help in positioning products effectively and evaluating the pricing competitiveness compared to competitors.

4. **Sales Channels**: In-store retail sales outperform other sales methods, indicating that consumers prefer the physical shopping experience. Strengthening in-store promotions and experiences can lead to increased sales and brand loyalty.

5. **Sales Performance Metrics**: Total sales show a strong positive correlation with operating profit, price per unit, and units sold. This highlights the importance of optimizing these metrics to drive revenue growth.

6. **Regional Impact**: Sales differ across regions, suggesting that regional characteristics and preferences influence purchasing decisions. Tailored marketing strategies for different regions can enhance customer engagement and sales.

7. **Retailer Influence**: Retailers also play a significant role in total sales. Collaborating closely with retailers and providing them with effective sales support can yield mutually beneficial outcomes.

In conclusion, these insights emphasize the need for Adidas to focus on product differentiation, price optimization, and targeted marketing to capitalize on high-demand products, boost in-store experiences, and strengthen collaborations with retailers. By leveraging these conclusions, Adidas can make informed decisions to drive revenue growth and maintain a competitive edge in the market. As a data-driven organization, adopting these insights will empower Adidas to navigate market dynamics effectively and sustain long-term success.
